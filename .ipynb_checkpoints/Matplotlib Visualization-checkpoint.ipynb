```python
import matplotlib.pyplot as plt

%matplotlib inline

import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
plt.plot(x,y,'r');
plt.xlabel('X-Axis Title')
plt.ylabel('Y-Axis Title')
plt.title('My Visualization Title');
```

### creating multiple plots in same canvas

```python
plt.subplot(1,3,1) # nrows, ncols, plot_number

plt.plot(x,y,'r--')

plt.subplot(1,3,2)

plt.plot(y,x,'g');

plt.subplot(1,3,3)

plt.plot(y,x,'b');
```

### Object Oriented Method

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8]) # left,bottom,width,height

axes.plot(x,y,'r');
axes.set_xlabel('X-Axis Label')
axes.set_ylabel('Y-Axis Label')
axes.set_title('Title')
```

```python
fig = plt.figure(figsize=(18,9))

axes1 = fig.add_axes([0.1,0.1,0.9,0.9]) # left,bottom,width,height
axes2 = fig.add_axes([0.2,0.6,0.4,0.3]) # left,bottom,width,height

# plotting for larger frame
axes1.plot(x,y,'b')

# plotting for small aer frame
axes2.plot(y,x,'r')
```

## Displaying legends

```python
fig = plt.figure(figsize=(12,6))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="x**2")
ax.plot(x,x**3,label="x**3")
ax.legend();#loc=4);
ax.legend(loc=1) # upper right corner
ax.legend(loc=2) # upper left corner
ax.legend(loc=3) # lower left corner
ax.legend(loc=4) # lower right corner
```

```python
fig = plt.figure(figsize=(12,6))

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label="x**2",linewidth=2.5,linestyle='-',marker='o')

ax.plot(x,
        x**3,
        label="x**3",
        linewidth=3.5,
        linestyle="-.",
        marker='s',
        markersize=9,
        markerfacecolor='yellow',
       markeredgecolor='blue',
       markeredgewidth=3)



ax.legend();#loc=4);


```
